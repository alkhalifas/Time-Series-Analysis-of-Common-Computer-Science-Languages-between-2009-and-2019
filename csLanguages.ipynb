```python
import numpy as pn
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

import os
import time
```

```python
df = pd.read_csv("csl_multiTimeline.csv")
```

```python
df.head()
```

```python
df.columns = ['Month','Python','SQL','R','Java','C']
```

```python
df.Month = pd.to_datetime(df.Month)
df.set_index('Month', inplace=True)
```

```python
df.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
df[['Python']].plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

```python
Python = df[['Python']]
R = df[['R']]
SQL = df[['SQL']]
C = df[['C']]
Java = df[['Java']]
X=7
```

```python
df_rm = pd.concat([Python.rolling(X).mean(), R.rolling(X).mean(), SQL.rolling(X).mean(), C.rolling(X).mean(), Java.rolling(X).mean()], axis = 1)
df_rm.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Popularity (%)', fontsize=20);
```
