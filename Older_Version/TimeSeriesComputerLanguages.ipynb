```python
# Import your packages and data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()

df = pd.read_csv("multiTimeline_2.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Rename the columns of your data frame so there are no white spaces in them:
df.columns = ['Week','Python','SQL','R','C'] #List

df.Week = pd.to_datetime(df.Week)
df.set_index('Week', inplace=True)
```

```python
df.head()
```

```python
# Exploratory Data Analysis:

df.plot(figsize=(35,10), linewidth=3, fontsize=20)

plt.xlabel('Years', fontsize=40)
plt.title("Programming Languages Popularity over 5 years", fontsize=40)
```

```python
# Trends and Seasonality in the time series data

python = df[['Python']] # two brackets makes it a dataframe, one will make it a pandas series
python.rolling(3).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)

r = df[['R']] # two brackets makes it a dataframe, one will make it a pandas series
r.rolling(3).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)

sql = df[['SQL']] # two brackets makes it a dataframe, one will make it a pandas series
sql.rolling(3).mean().plot(figsize=(20,10), linewidth=5, fontsize=20)

plt.xlabel('Year', fontsize=20);
```

```python
df_rm = pd.concat([python.rolling(4).mean(), r.rolling(4).mean(), sql.rolling(4).mean()], axis=1)
df_rm.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Search Popularity', fontsize=20)
```
