## Quick Time Series Analysis of Computer Science Languages from Google Trends

The purpose of this page is to do a quick time-series analysis of historical records of google searches of various languages to indicate the change in their popularity over time. We hypothesize that the popularity of python and SQL will increase over this time period. We also hypothesize that the google-search popularity will fall during the winter and 4th of July holidays. We will utilize Google Trends to find our data, and we will then use a rolling average to reduce noise. We will need to run a few different values to ensure that our data of interest is not averaged out.

Import libraries of interest

```python
import numpy as pn
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

import os
import time
```

Import the dataset

```python
# Dataset obtained from Google Trends
df = pd.read_csv("csl_multiTimeline.csv")
```

Check your data

```python
df.head()
```

Define the columns of interest

```python
df.columns = ['Month','Python','SQL','R','Java','C']
```

```python
df.Month = pd.to_datetime(df.Month)
df.set_index('Month', inplace=True)
```

Plot the raw data

```python
df.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

Plot "Python" Alone

```python
df[['Python']].plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20);
```

Define the languages as lists

```python
Python = df[['Python']]
R = df[['R']]
SQL = df[['SQL']]
C = df[['C']]
Java = df[['Java']]
X=7
```

Use a rolling average to smooth your data and eliminate noise, and then plot it

```python
df_rm = pd.concat([Python.rolling(X).mean(), R.rolling(X).mean(), SQL.rolling(X).mean(), C.rolling(X).mean(), Java.rolling(X).mean()], axis = 1)
df_rm.plot(figsize=(20,10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Popularity (%)', fontsize=20);
```

Reduce the rolling average, and replot your data. We notice there is a trend around the winter holidays

```python
X=1
df_rm = pd.concat([Python.rolling(X).mean(), R.rolling(X).mean(), SQL.rolling(X).mean()], axis = 1)
df_rm.plot(figsize=(20,10), linewidth=5, fontsize=20)

plt.xlabel('Year', fontsize=20)
plt.ylabel('Popularity (%)', fontsize=20)


plt.axvline(pd.to_datetime('2016-01-01'), color='r', linestyle='--', lw=2)
```

Plot vertical lines to show the trends during the holidays more clearly, replot

```python
X=1
df_rm = pd.concat([Python.rolling(X).mean(), R.rolling(X).mean(), SQL.rolling(X).mean()], axis = 1)
df_rm.plot(figsize=(20,10), linewidth=5, fontsize=20)

plt.xlabel('Year', fontsize=20)
plt.ylabel('Popularity (%)', fontsize=20)


plt.axvline(pd.to_datetime('2012-01-01'), color='r', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2013-01-01'), color='r', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2014-01-01'), color='r', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2015-01-01'), color='r', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2016-01-01'), color='r', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2017-01-01'), color='r', linestyle='--', lw=2)
```

Conclusion: We were able to show that the popularity of Python has infact increase dramatically over the last few years as predicted. SQL however seems to show an opposite trend. We used a rolling average to see this trend by using a value of X=1. Our next steps are to explore further and attempt to forecase changes in popularity over the next decade.
